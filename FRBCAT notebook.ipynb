# Input data from FRBCAT

```python
t = Table.read('frbcat_20180914_confirmed.csv', format='ascii.csv')
```

```python
# Print column names
t.columns.keys()
```

## Convert hoursangle to degrees (and then to radian)

```python
c = SkyCoord(t['rop_raj'], t['rop_decj'], frame='icrs', unit=(u.hourangle, u.deg))
```

```python
ra_rad = c.ra.wrap_at(180 * u.deg).radian
dec_rad = c.dec.radian
```

# Plot FRB coordinates

```python
plt.figure(figsize=(8,4.2))
ax = plt.subplot(111, projection="aitoff")
ax.grid(True)

# dms = [float(item.split('&')[0]) for item in t['rmp_dm']]
# norm = plt.Normalize()
# colors = plt.cm.hot(norm(dms))

ax.plot(ra_rad, dec_rad, 'o', markersize=3, alpha=0.9)
# ax.set_xticklabels(['14h','16h','18h','20h','22h','0h','2h','4h','6h','8h','10h'])
plt.subplots_adjust(top=0.95,bottom=0.0)
plt.show()
```
