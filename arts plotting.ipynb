```python
import numpy as np
from matplotlib import pyplot as plt
import h5py
import glob 
```

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
# Set font size for labels and axes ticks (http://matplotlib.org/users/customizing.html)
from matplotlib import rc
rc('font', size=12)
rc('axes', titlesize=14)
rc('axes', labelsize=14)
```

# Read data

```python
#flist = glob.glob('/data/03/Triggers/triggers/CB21/*hdf5')
flist = glob.glob('data/*hdf5')
```

```python
flist
```

```python
np.sort(flist)
```

```python
f = h5py.File(np.sort(flist)[2], 'r')
print(f.items())

data_freq = f['data_freq_time'][:]
data_dm = f['data_dm_time'][:]
```

# Plot results

```python
plt.imshow(data_freq, aspect='auto', vmax=3, vmin=-3)
plt.xlabel('Time')
plt.ylabel('Freq')
plt.colorbar()
plt.show()
```

```python
plt.imshow(data_dm, aspect='auto')
plt.xlabel('Time')
plt.ylabel('DM')
plt.colorbar()
plt.show()
```
